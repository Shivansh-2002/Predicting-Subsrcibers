```python
import pandas as pd

X_train = pd.read_csv('../input/kaggle-knight1/xtrain_csv.csv')

X_train.head()
```

```python
X_train = X_train.drop("Unnamed: 0",axis = 1)
X_train = X_train.drop("default",axis = 1)
X_train = X_train.drop("month",axis = 1)
X_train = X_train.drop("poutcome",axis = 1)
X_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
str_cols =  ['job','marital','education','housing','loan','contact']
X_train[str_cols] = X_train[str_cols].apply(LabelEncoder().fit_transform)
X_train.head()
```

```python
y_train = pd.read_csv('../input/kaggle-knight1/ytrain_csv.csv')
y_train.head()
```

```python
y_train = y_train.drop("Id",axis = 1)
y_train.head()
```

```python
y_train.Category = LabelEncoder().fit_transform(y_train.Category)

y_train.head()
```

```python
X_test_data = pd.read_csv('../input/kaggle-knight1/xtest_csv.csv')
X_test = X_test_data
X_test = X_test.drop("Unnamed: 0",axis = 1)
X_test = X_test.drop("default",axis = 1)
X_test = X_test.drop("month",axis = 1)
X_test = X_test.drop("poutcome",axis = 1)
test_str_cols =  ['job','marital','education','housing','loan','contact']
X_test[test_str_cols] = X_test[str_cols].apply(LabelEncoder().fit_transform)
X_test.head()
```
